### data manipulation

```python
# 1. renaming columns
df.rename({'details':'configuration','title':'product_name'}, axis=1, inplace= True)
```

```python
df
```

```python
# 2. adding column
df['website'] = 'flipkart'
```

```python
df
```

```python
# fun to extract rating
import numpy as np

def extract_rating(value):
    if value:
        value_0 = value.split()[0]
        value_0 = value_0.replace(',','')
        return int(value_0)
    else:
        return np.nan
# adding columns with functions
df['ratings'] = df['rr'].apply(extract_rating)

```

```python
def extract_reviews(val):
    if val:
        val_1 = val.split()[-2]
        val_1 = val_1.replace(',','')
        return int(val_1)
    else:
        return np.nan

df['reviews'] = df['rr'].apply(extract_reviews)
```

```python
df
```

```python
# 4. dropping columns
df.drop(columns=['rr'], inplace=True)
```

```python
df
```

```python
# Task 1: create a column brand and store the brand name in it taken from the first column
df['brand'] = df['product_name'].apply(lambda x: x.split()[0])
```

```python
df
```

```python
# 5. fixing the columns values for example price
df['price'] = df['price'].apply(lambda x: x.replace('₹','').replace(',',''))
```

```python
df
```

```python
df['link'] = df['link'].apply(lambda x : 'https://www.flipkart.com'+x)
```

```python
df['price'] = df['price'].astype(int)
```

```python
df.info()
```

```python
df['brand'].value_counts()
```

```python
df['brand'].value_counts().plot(kind='bar')
```

```python
df['price'].plot(kind='hist', bins=25, color='red', alpha=0.5, title= 'Price distribution')
```

```python
df['reviews'].plot(kind='line',title='Reviews',figsize=(15,5))
```

```python
# 6.sorting the data by column
df.sort_values(by='price', ascending=False, inplace=True)
```

```python
df.sort_values(by=['reviews','ratings'], ascending=False, inplace=True)
```

```python
df
```

```python
# 7. remove duplicate records
df.drop_duplicates(subset=['product_name'],inplace=True)
```
